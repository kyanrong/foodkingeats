```python
import csv
import pandas as pd

df = pd.read_csv('places.csv')
df['address'] = df['address_street']
for index, row in df.iterrows():
  address = "\"street\":\"{street}\", \"unit\":\"{unit}\", \"building\":\"{building}\", \"postal\":\"{postal}\""
  df.at[index, 'address'] = address.format(street=row['address_street'], unit=row['address_unit'], building=row['address_building'], postal=str(row['address_postal']))
df = df.drop(columns=['address_street', 'address_unit', 'address_building', 'address_postal'])

df['id'] = df['id'].apply(lambda x: "('" + x + "'")
df['name'] = df['name'].apply(lambda x: "'" + x + "'")
df['isHalal'] = df['isHalal'].apply(lambda x: str(x).upper() + "),")
df['address'] = df['address'].apply(lambda x: "'{" + x + "}'")

df['final'] = df['id'] + "," + df['name'] + "," + df['address'] + "," + df['isHalal']

df = df.drop(columns=['id', 'name', 'isHalal', 'address'])

df.to_csv('places-cleaned.csv', index=None, header=None)

df.head()
```

```python
import pandas as pd

df = pd.read_csv('visits.csv')
df['id'] = df['id'].apply(lambda x: "('" + x + "'")
df['PlaceId'] = df['PlaceId'].apply(lambda x: "'" + x + "'")
df['isSponsored'] = df['isSponsored'].apply(lambda x: str(x).upper())
df['rating'] = df['rating'].apply(lambda x: str(x).upper())
df['youtubeUrl'] = df['youtubeUrl'].apply(lambda x: "'" + x + "'")
df['dateOfUpload'] = df['dateOfUpload'].apply(lambda x: "'" + x + "'),")

df['final'] = df['id'] + "," + df['PlaceId'] + "," + df['isSponsored'] + "," + df['rating'] + "," + df['youtubeUrl'] + "," + df['dateOfUpload']

df = df.drop(columns=['id', 'PlaceId', 'isSponsored', 'rating', 'youtubeUrl', 'dateOfUpload'])

df.to_csv('visits-cleaned.csv', index=None, header=None)

df.head()
```

```python
import pandas as pd

df = pd.read_csv('food.csv')

df.fillna('null', inplace=True)

df['PlaceId'] = df['PlaceId'].apply(lambda x: "('" + x + "'")
df['VisitId'] = df['VisitId'].apply(lambda x: "'" + x + "'")
df['hasPriceRange'] = df['hasPriceRange'].apply(lambda x: str(x).upper())
df['priceMin'] = df['priceMin'].apply(lambda x: str(x))
df['priceMax'] = df['priceMax'].apply(lambda x: str(x))
df['notes'] = df['notes'].apply(lambda x: "'" + x + "'")
df['name'] = df['name'].apply(lambda x: "'" + x + "'")
df['currency'] = df['currency'].apply(lambda x: "'" + x + "'),")

df['final'] = df['PlaceId'] + "," + df['VisitId'] + "," + df['hasPriceRange'] + "," + df['priceMin'] + "," + df['priceMax'] + "," + df['notes'] + "," + df['name'] + "," + df['currency']

df = df.drop(columns=['PlaceId', 'VisitId', 'hasPriceRange', 'priceMin', 'priceMax', 'notes', 'name', 'currency'])

df.to_csv('food-cleaned.csv', index=None, header=None)

df.head()
```
